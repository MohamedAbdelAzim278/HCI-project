```python
def paint():
    class main:
        def __init__(self,master):
            self.master = master
            self.color_fg = 'black'
            self.color_bg = 'white'
            self.old_x = None
            self.old_y = None
            self.penwidth = 5
            self.drawWidgets()
            self.c.bind('<B1-Motion>',self.paint)#drwaing the line 
            self.c.bind('<ButtonRelease-1>',self.reset)

        def paint(self,e):
            if self.old_x and self.old_y:
                self.c.create_line(self.old_x,self.old_y,e.x,e.y,width=self.penwidth,fill=self.color_fg,capstyle=ROUND,smooth=True)

            self.old_x = e.x
            self.old_y = e.y

        def reset(self,e):    #reseting or cleaning the canvas 
            self.old_x = None
            self.old_y = None      

        def changeW(self,e): #change Width of pen through slider
            self.penwidth = e
           
        def clear(self):
            self.c.delete(ALL)

        def change_fg(self):  #changing the pen color
            self.color_fg=colorchooser.askcolor(color=self.color_fg)[1]

        def change_bg(self):  #changing the background color canvas
            self.color_bg=colorchooser.askcolor(color=self.color_bg)[1]
            self.c['bg'] = self.color_bg

        def drawWidgets(self):
            self.controls = Frame(self.master,padx = 5,pady = 5)
            Label(self.controls, text='Pen Width:',font=('arial 18')).grid(row=0,column=0)
            self.slider = ttk.Scale(self.controls,from_= 5, to = 100,command=self.changeW,orient=VERTICAL)
            self.slider.set(self.penwidth)
            self.slider.grid(row=0,column=1,ipadx=30)
            self.controls.pack(side=LEFT)
        
            self.c = Canvas(self.master,width=500,height=400,bg=self.color_bg,)
            self.c.pack(fill=BOTH,expand=True)

            menu = Menu(self.master)
            self.master.config(menu=menu)
            filemenu = Menu(menu)
            colormenu = Menu(menu)
            menu.add_cascade(label='Colors',menu=colormenu)
            colormenu.add_command(label='Brush Color',command=self.change_fg)
            colormenu.add_command(label='Background Color',command=self.change_bg)
            optionmenu = Menu(menu)
            menu.add_cascade(label='Options',menu=optionmenu)
            optionmenu.add_command(label='Clear Canvas',command=self.clear)
            optionmenu.add_command(label='Exit',command=self.master.destroy) 
        
    if __name__ == '__main__':
        root = Tk()
        main(root)
        root.title('Application')
        root.mainloop()
```

```python
def pingpong():
    # Initialize the game
    pygame.init()

    # Set up the game window
    window_width = 800
    window_height = 400
    window = pygame.display.set_mode((window_width, window_height))
    pygame.display.set_caption("Ping Pong Game")

    # Set up the colors
    BLACK = (0, 0, 0)
    WHITE = (255, 255, 255)

    # Set up the paddles
    paddle_width = 10
    paddle_height = 60
    paddle_speed = 5
    player_paddle_x = window_width - paddle_width - 10
    player_paddle_y = window_height // 2 - paddle_height // 2
    ai_paddle_x = 10
    ai_paddle_y = window_height // 2 - paddle_height // 2

    # Set up the ball
    ball_radius = 10
    ball_speed_x = -3
    ball_speed_y = 3
    ball_x = window_width // 2
    ball_y = window_height // 2

    # Set up the game clock
    clock = pygame.time.Clock()

    # Game loop
    running = True
    while running:
        # Handle events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Move the player paddle based on mouse position
        mouse_x, mouse_y = pygame.mouse.get_pos()
        player_paddle_y = mouse_y - paddle_height // 2

        # Move the AI paddle to track the ball
        if ball_speed_x < 0 and ball_x < window_width // 2:
            if ai_paddle_y < ball_y:
                ai_paddle_y += paddle_speed
            elif ai_paddle_y > ball_y:
                ai_paddle_y -= paddle_speed

        # Update the ball position
        ball_x += ball_speed_x
        ball_y += ball_speed_y

        # Check collision with the paddles
        if (ball_x >= player_paddle_x and player_paddle_y <= ball_y <= player_paddle_y + paddle_height) or \
                (ball_x <= ai_paddle_x + paddle_width and ai_paddle_y <= ball_y <= ai_paddle_y + paddle_height):
            ball_speed_x *= -1

        # Check collision with walls
        if ball_y <= 0 or ball_y >= window_height - ball_radius:
            ball_speed_y *= -1

        # Check if the ball goes out of bounds
        if ball_x < 0 or ball_x > window_width:
            # Reset ball position
            ball_x = window_width // 2
            ball_y = window_height // 2
            # Randomize ball direction
            ball_speed_x *= random.choice([1, -1])
            ball_speed_y *= random.choice([1, -1])

        # Clear the window
        window.fill(BLACK)

        # Draw the player paddle
        pygame.draw.rect(window, WHITE, (player_paddle_x, player_paddle_y, paddle_width, paddle_height))

        # Draw the AI paddle
        pygame.draw.rect(window, WHITE, (ai_paddle_x, ai_paddle_y, paddle_width, paddle_height))

        # Draw the ball
        pygame.draw.circle(window, WHITE, (ball_x, ball_y), ball_radius)

        # Update the display
        pygame.display.flip()

        # Set the frame rate
        clock.tick(60)

    # Quit the game
    pygame.quit()
```

```python
import cv2
import numpy as np
import pyautogui
from tkinter import messagebox
import pygame
import random
from tkinter import *
from tkinter import ttk, colorchooser

messagebox.showinfo("Hello", "Choose one of the mystery games by hovering on any of the green boxes!")

# Define the button coordinates and size
button1_x = 100
button1_y = 300
button1_width = 100
button1_height = 50

button2_x = 400
button2_y = 300
button2_width = 100
button2_height = 50

cap = cv2.VideoCapture(0)

while True:
    # Take each frame
    ret, frame = cap.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    lower_red = np.array([0, 150, 150])
    upper_red = np.array([255, 255, 255])
    mask = cv2.inRange(hsv, lower_red, upper_red)
    (minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(mask)

    if maxVal > 0:
        # Extract the x and y coordinates of the laser
        laser_x, laser_y = maxLoc
        
        # Get the screen size
        screen_width, screen_height = pyautogui.size()

        # Map the laser position to the screen coordinates
        mouse_x = int(laser_x * screen_width / frame.shape[1])
        mouse_y = int(laser_y * screen_height / frame.shape[0])

        # Move the mouse
        pyautogui.moveTo(mouse_x, mouse_y)

        # Check if the laser is pointing at the button
        if button1_x < laser_x < button1_x + button1_width and button1_y < laser_y < button1_y + button1_height:
            # Perform the action when the laser points at the button
            messagebox.showinfo("Button Pressed", "Lets play pingpong!")
            game = 1
            break
        elif button2_x < laser_x < button2_x + button2_width and button2_y < laser_y < button2_y + button2_height:
            messagebox.showinfo("Button Pressed", "Lets play paint!")
            game = 2
            break
            # Add your desired action here

    # Draw the button on the frame
    cv2.rectangle(frame, (button1_x, button1_y), (button1_x + button1_width, button1_y + button1_height), (255, 255, 255), 2)
    cv2.rectangle(frame, (button2_x, button2_y), (button2_x + button2_width, button2_y + button2_height), (255, 255, 255), 2)
    
    cv2.circle(frame, maxLoc, 20, (255, 0, 0), 2, cv2.LINE_AA)
    cv2.imshow('Track Laser', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()

if game == 1:
    pingpong()
if game == 2:
    paint()    
```
